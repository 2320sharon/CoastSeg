```python
shorelines_folder=os.path.abspath(os.getcwd())+os.sep+"Coastseg"+os.sep+"shorelines"
shorelines=glob.glob(shorelines_folder+os.sep+"*.geojson")
print("shorelines: ",shoreline,"\n")
shoreline_layer_names=[]
for shoreline in shorelines:
    shoreline_layer_name=os.path.basename(shoreline)
    shoreline_layer_names.append(shoreline_layer_name)
shoreline_total_bounds=[]
for shoreline_file in shorelines:
    transects_layer_name=os.path.splitext(os.path.basename(shoreline_file))[0]
    print(transects_layer_name)
    shoreline_gpd=bbox.read_gpd_file(shoreline_file)
    print(shoreline_gpd.total_bounds)
    bounds=shoreline_gpd.total_bounds
    print(bounds,"\n")
    shoreline_total_bounds.append(bounds)
```

```python
shoreline_total_bounds
```

```python
df = pd.DataFrame( shoreline_total_bounds,columns=['minx', 'miny', 'maxx', 'maxy'])
# insert a new column with the names of each geojson file
df.insert(0, 'filename', shoreline_layer_names)
df
df.to_csv("shorelines_bounding_boxes.csv",index=False)
```
